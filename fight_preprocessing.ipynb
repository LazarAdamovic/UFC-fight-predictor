```python
import pandas as pd

```

```python
# read fight night csv

df=pd.read_csv('fighter_scraper/fightnightpandas.csv')

df
```

New 

```python
pd.read_html(df['link'][0])[0]
```

```python


data = pd.DataFrame()

for index, row in df.iterrows():
    # this gets the table information from the url
    tempdf = pd.read_html(row['link'])[0]
    # clean the table
    tempdf=tempdf[['W/L','Fighter','Method','Round','Time']]
    # combine the table with scraped information
    tempdf['location']=row['location']
    tempdf['date']=row['date']
    # concat
    data = pd.concat([data,tempdf]) 
```

```python
# count unique fights

data.nunique()
```

```python
data.isnull().sum()
```

```python
data=data.dropna(axis=0, how='any')
```

```python
data[['Method']]
```

```python
# get rid of time column - Round is bascially a binned version of time which is more readable by algorithms

data=data[['W/L', 'Fighter', 'Method', 'Round', 'location', 'date']]
```

```python
# save to csv

data.to_csv('fights-all.csv', index=False)
```
